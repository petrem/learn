```python
goats = ["goat1", "goat2"]
doors = goats + ["car"]

def choose_goat():
    chosen = np.random.choice(goats)
    other = goats[1 - goats.index(chosen)]
    return [chosen, other]

outcomes_by_choice = {
    "goat1": lambda: ["goat1", "goat2", "car"],
    "goat2": lambda: ["goat2", "goat1", "car"],
    "car": lambda: ["car"] + choose_goat(),
}

def make_dataframe(data=None):
    return pd.DataFrame(data=data, columns=("guessed", "revealed", "remained"))

def monty_hall_game():
    """returns ["guess", "revealed", "remains"]"""
    choice = np.random.choice(choices)
    return make_dataframe([outcomes_by_choice[choice]()])
    
```

```python
_trials = [monty_hall_game() for _ in range(10000)]
trials = make_dataframe().append(_trials, ignore_index=True)
trials
```
