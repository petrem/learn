```python
a = np.arange(11, 50)
```

```python
def teen(x):
    return 13 <= x <= 19
```

```python
ages = pd.DataFrame({"Age": a})
#ages = ages.with_column('Teenager', ages.apply(teen, 'Age'))
ages["Teenager"] = ages.applymap(teen)
ages
```

```python
sum(ages["Teenager"])
```

```python
still_young = lambda x: x >=35 and x < 40
```

```python
ages.where(ages["Age"].apply(still_young)).count()
```

```python
coin = np.array(('heads', 'tails'))
np.random.choice(coin)
```

```python
sum(np.random.choice(coin, 100) == 'heads')
```

```python
num_heads=np.array([])
for i in np.arange(10000):
    num_heads = np.append(num_heads, sum(np.random.choice(coin, 100) == 'heads'))
num_heads
```

```python
t=pd.DataFrame({"#heads in 100 tosses": num_heads})
t
```

```python
 t.hist(bins=np.arange(30, 70, 1))
```

```python
n=t["#heads in 100 tosses"]
n.where(np.logical_and(n>40, n < 60)).count() / 10000
```
